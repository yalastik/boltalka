```python
def clear_messages(filename=FILE_NAME):
    with open(filename, 'r') as f:
        dialogs = f.read()
    messages = dialogs.split('\n')
    translator = str.maketrans('', '', string.punctuation)
    data = []
    for message in messages:
        data.append(message.translate(translator).lower().split())
    return data
```

```python
def clear_message(message):
    translator = str.maketrans('', '', string.punctuation)
    return message.translate(translator).lower().split()
```

```python
def get_messages(filename=FILE_NAME):
    with open(filename, 'r') as f:
        dialogs = f.read()
    return dialogs.split('\n')
```

```python
class Sentence():
    def __init__(self, sentence):
        self.sentence = sentence
        
        
    def add_vector(self, vector):
        self.vector = vector

        
    def find_similarity(self, vector):
        return 1 - cosine(self.vector, vector)
```

```python
def create_voc(data, model):
    voc = []
    for sent in data:
        sent_obj = Sentence(sent)
        sent_obj.add_vector(find_vector(model,clear_message(sent)))
        voc.append(sent_obj)
    return voc
```

```python
def find_vector(model, message):
    vector = np.zeros(model.wv.vector_size)
    for word in message:
        if word in model.wv.vocab:
            vector = vector + model.wv[word]
    return vector
```

```python
def n_most_similar(model, message, n, m2v_model=m2v):
    vector = find_vector(model, clear_message(message))
    heap = []
    for i, sent in enumerate(m2v_model.vectors):
        sim = cosine(sent.vector, vector)
        heapq.heappush(heap,(sim,i))
    return heap[:n]
```
